# Functoin for Five Folds extraction from Dataset
    Parameters:instance- A Pandas data frame containing the instances
        
    Returns: fold_0, fold_1, fold_2, fold_3, fold_4 having equal distribuitiom
        
   

```python
def extract_k_fold(instance):
    instance = instance.sample(frac=1).reset_index(drop=True) # mix dataset randomly
    columns_no = len(instance.columns) # number of columns
    rows_no = len(instance.index) # number of rows
    # initiationg 5 empty folds i.e. pandas dataframe
    fold_0 = pd.DataFrame(columns=(instance.columns))
    fold_1 = pd.DataFrame(columns=(instance.columns))
    fold_2 = pd.DataFrame(columns=(instance.columns))
    fold_3 = pd.DataFrame(columns=(instance.columns))
    fold_4 = pd.DataFrame(columns=(instance.columns))
    correct_class_column = columns_no - 1 # Last column defines the actual class
    # generate an array containing the unique  actual class values
    total_unique_classes = instance.iloc[:,correct_class_column].sort_values().unique()
    total_unique_classes_list_np_size = total_unique_classes.size
    # For each unique class in the unique Actual Class array
    for unique_class_list_np_idx in range (total_unique_classes_list_np_size):
        count = 0
        #distributing eqally instances of unique classes over the 5 folds
        for row in range(rows_no):
            if total_unique_classes[unique_class_list_np_idx] == (
                instance.iloc[row,correct_class_column]):
                    # fold_0
                    if count == 0:
                        n_r = instance.iloc[row,:]
                        fold_0.loc[len(fold_0)] = n_r
                        count += 1
                    #fold_1
                    elif count == 1:
                        n_r = instance.iloc[row,:]
                        fold_1.loc[len(fold_1)] = n_r                                    
                        count += 1
                    #fold_2
                    elif count == 2:
                        n_r = instance.iloc[row,:]
                        fold_2.loc[len(fold_2)] = n_r
                        count += 1
                    #fold_3
                    elif count == 3:
                        n_r = instance.iloc[row,:]
                        fold_3.loc[len(fold_3)] = n_r
                        count += 1
                    #fold_4
                    else:
                        n_r = instance.iloc[row,:]
                        fold_4.loc[len(fold_4)] = n_r
                        count= 0
         
    return fold_0, fold_1, fold_2, fold_3, fold_4
```

# Sigmoid Function
Parameters:
        z: A real number
Returns: 
        1.0/(1 + np.exp(-z))

```python
def sigmoid(x):
    return 1.0/(1 + np.exp(-x))
 
```

# Gradient Descent
parameters:
      trainingset
      
      
Returns:
      weights-vector of weights(theta)

```python
def gradient_descent(training_set):
    
    no_of_columns = training_set.shape[1]
    no_of_rows = training_set.shape[0]
    x = training_set[:,:(no_of_columns- 1)]#getting the attributes from the training set.
    no_of_attributes = x.shape[1]
    actual_class = training_set[:,(no_of_columns - 1)]# actual_class is a 1d array.
    learning_rate = 0.01 #  learning rate can be varied
    maxiterations = 10000 # maximum number of iterations
    iter = 0
    exceeded_max_iter = False
    # set the tolerance. When the euclidean norm of the gradient vector (i.e. magnitude of the changes in the weights) gets below this value, 
    # stop iterating through the while loop
    tolarence = 0.001
    norm_of_gradient = None
    # Set a flag to determine if we have reached the minimum of the cost (i.e. error) function.
    converged = False
    # initialise the weights vector with random floats between -0.01 and 0.01
    weights = np.random.uniform(-0.01,0.01,(no_of_attributes))
    changes_in_weights = None
    # Keep running the loop below until convergence on the minimum of the 
    # cost function or we exceed the max number of iterations
    while(not(converged) and not(exceeded_max_iter)):
        changes_in_weights = np.zeros(no_of_attributes)
        for inst in range(no_of_rows):
            # weighted sum of the attributes for this instance
            output = np.dot(weights, x[inst,:])                
            y =  sigmoid(output)#  sigmoid of the weighted sum
            difference = (actual_class[inst] - y) # Calculate difference
            product = np.multiply(x[inst,:], difference)# multiplying the difference by the attribute vector
            changes_in_weights = np.add(changes_in_weights,product)# update the weight changes 
        step_size = np.multiply(changes_in_weights, learning_rate)
        weights = np.add(weights, step_size)
        # Test to see if we have converged on the minimum of the error
        # function
        norm_of_gradient = np.linalg.norm(changes_in_weights)
        if (norm_of_gradient < tolarence):
            converged = True
        iter += 1
        # if exceeded the maximum number of iterations
        if (iter > maxiterations):
            exceeded_max_iter = True
    return weights
 
```

# Logistic Regression
    Parameters:
      training_set- The training instances as a Pandas dataframe
      test_set- The test instances as a Pandas dataframe
      
   
    Returns:
      accuracy- classification accuracy as a decimal
      predictions- classifications of all the test instances 
      weights_for_each_class- weight vectors for each class 
      no_of_instances_test- number of test instances

```python
def logistic_regression(training_set, test_set):
    training_set = training_set.drop(
        training_set.columns[[0]], axis=1)
    test_set = test_set.drop(
        test_set.columns[[0]], axis=1)
    list_of_unique_classes = pd.unique(training_set["Actual Class"])# Make a list of the unique classes
    for cl in range(len(list_of_unique_classes)):
        training_set["Actual Class"].replace(
            list_of_unique_classes[cl], cl ,inplace=True)
        test_set["Actual Class"].replace(
            list_of_unique_classes[cl], cl ,inplace=True)
    # Insert a column of 1s in column 0 of both the training
    # and test sets. This is the bias and helps with gradient
    # descent. (i.e. X0 = 1 for all instances)
    training_set.insert(0, "Bias", 1)
    test_set.insert(0, "Bias", 1)
    np_training_set = training_set.values
    np_test_set = test_set.values
    # Add 2 additional columns to the testing dataframe
    test_set = test_set.reindex(
        columns=[*test_set.columns.tolist(
        ), 'Predicted Class', 'Prediction Correct?'])
 
                                    #Training Phase 
 
    no_of_columns_training_set = np_training_set.shape[1]
    no_of_rows_training_set = np_training_set.shape[0]
    trainingsets = []
    for cl in range(0, len(list_of_unique_classes)):
        temp = np.copy(np_training_set)
        # This class becomes the positive class 1  and other classes the negative class 0
        for row in range(no_of_rows_training_set):
            if (temp[row, (no_of_columns_training_set - 1)]) == cl:
                temp[row, (no_of_columns_training_set - 1)] = 1
            else:
                temp[row, (no_of_columns_training_set - 1)] = 0
        trainingsets.append(temp)
    weights_for_each_class = []
 
    for cl in range(0, len(list_of_unique_classes)):
        weights_for_this_class = gradient_descent(trainingsets[cl])
        weights_for_each_class.append(weights_for_this_class)
                                    # Testing Phase 
    no_of_columns_test_set = np_test_set.shape[1]
    no_of_rows_test_set = np_test_set.shape[0]
    x = np_test_set[:,:(no_of_columns_test_set - 1)]
    no_of_attributes = x.shape[1]
    actual_class = np_test_set[:,(no_of_columns_test_set - 1)]
    for inst in range(0,  no_of_rows_test_set):
        # Create a scorecard that keeps track of the probabilities of this
        # instance being a part of each class
        scorecard = []
        # Calculate and store the probability for each class in the scorecard
        for cl in range(len(list_of_unique_classes)):
            # Calculate weighted sum of the attributes for
            # this instance
            output = np.dot(weights_for_each_class[cl], x[inst,:])
            # Calculate the sigmoid of the weighted sum
            # This is the probability that this instance belongs
            # to the positive class
            this_probability = sigmoid(output)
            scorecard.append(this_probability)
        most_likely_class = scorecard.index(max(scorecard))
        test_set.loc[inst, "Predicted Class"] = most_likely_class
        # Update the 'Prediction Correct?' column of the test_set data frame
        # 1 if correct, else 0
        if test_set.loc[inst, "Actual Class"] == test_set.loc[
            inst, "Predicted Class"]:
            test_set.loc[inst, "Prediction Correct?"] = 1
        else:
            test_set.loc[inst, "Prediction Correct?"] = 0
    accuracy = (test_set["Prediction Correct?"].sum())/(len(test_set.index))
    predictions = test_set
    for cl in range(len(list_of_unique_classes)):
        predictions["Actual Class"].replace(
            cl, list_of_unique_classes[cl] ,inplace=True)
        predictions["Predicted Class"].replace(
            cl, list_of_unique_classes[cl] ,inplace=True)
    predictions['Prediction Correct?'] = predictions[
        'Prediction Correct?'].map({1: "Yes", 0: "No"})
    weights_for_each_class = pd.DataFrame(np.row_stack(weights_for_each_class))
    for cl in range(0, len(list_of_unique_classes)):
        row_name = str(list_of_unique_classes[cl] + " weights")        
        weights_for_each_class.rename(index={cl:row_name}, inplace=True)
    training_set_names = list(training_set.columns.values)
    training_set_names.pop() # Remove 'Actual Class'
    # Rename the column names
    for col in range(len(training_set_names)):
        col_name = str(training_set_names[col])        
        weights_for_each_class.rename(columns={col:col_name}, inplace=True)
    no_of_instances_test = len(test_set.index)
    return accuracy, predictions, weights_for_each_class, no_of_instances_test
```

# Driver Function

```python
seperator = ","  
def main():
    data_path = "preprocessed_data_set.csv"
    # Read the full text file and store records in a Pandas dataframe
    pd_data_set = pd.read_csv(data_path, sep=seperator)
    # Show functioning of the program
    trace_runs_file = "trace_runs_file.txt"
    outfile_tr = open(trace_runs_file,"w") 
    # Testing statistics
    test_stats_file = "test_stats_file.txt" 
    #test_stats_file = "iris_logistic_regression_test_stats.txt"
    # Open a test_stats_file 
    outfile_ts = open(test_stats_file,"w")
    # The number of folds in the cross-validation
    NO_OF_FOLDS = 5
    # Generate the five stratified folds
    fold_0, fold_1, fold_2, fold_3, fold_4 = extract_k_fold(pd_data_set)
    training_dataset = None
    test_dataset = None
    accuracy_statistics = np.zeros(NO_OF_FOLDS)
    # Run Logistic Regression the designated number of times as indicated by the number of folds
    for itr in range(0, NO_OF_FOLDS):
        print("Executing experimeent" + str(itr + 1) + "\n")
        print("\n")
        outfile_tr.write("Executing experiiment " + str(itr + 1) + "\n")
        outfile_tr.write("\n")
        if itr == 0:
            test_dataset = fold_0
            training_dataset = pd.concat([
               fold_1, fold_2, fold_3, fold_4], ignore_index=True, sort=False)                
        elif itr == 1:
            test_dataset = fold_1
            training_dataset = pd.concat([fold_0, fold_2, fold_3, fold_4], ignore_index=True, sort=False) 
        elif itr == 2:
            test_dataset = fold_2
            training_dataset = pd.concat([fold_0, fold_1, fold_3, fold_4], ignore_index=True, sort=False) 
        elif itr == 3:
            test_dataset = fold_3
            training_dataset = pd.concat([fold_0, fold_1, fold_2, fold_4], ignore_index=True, sort=False) 
        else:
            test_dataset = fold_4
            training_dataset = pd.concat([fold_0, fold_1, fold_2, fold_3], ignore_index=True, sort=False)     
        accuracyy, prediction, weight_for_each_class, no_of_instances_test = (
        logistic_regression(training_dataset,test_dataset))
        print("accuracy:")
        print(str(accuracyy * 100) +   "%")
        print("\n")
        print("Classification:\n")
        print(prediction)
        print("\n")
        print("Learned model:")
        print(weight_for_each_class)
        print("\n")
        print("Number of Test Instances")
        print(str(no_of_instances_test))
        print("\n") 
        outfile_tr.write("Accuracy:")
        outfile_tr.write(str(accuracyy * 100) + "%\n")
        outfile_tr.write("Classification:\n")
        outfile_tr.write(str(prediction) + "\n")
        outfile_tr.write("Learned model:\n")
        outfile_tr.write(str(weight_for_each_class) + "\n")
        outfile_tr.write("Number of Test Instances:")
        outfile_tr.write(str(no_of_instances_test) + "\n\n\n")
        accuracy_statistics[itr] = accuracyy
    outfile_ts.write(" summary Statistics\n")
    outfile_ts.write("----------------------------------------------------------\n")
    outfile_ts.write("Data Set: " + data_path + "\n")
    outfile_ts.write("\n")
    outfile_ts.write("accuracy statistics for All five Experiments:")
    outfile_ts.write(np.array2string(
        accuracy_statistics, precision=2, separator=',',
        suppress_small=True))
    outfile_ts.write("\n")
    outfile_ts.write("\n")
    accuracyy = np.mean(accuracy_statistics)
    accuracyy *= 100
    outfile_ts.write("Classification Accuracy : " + str(accuracyy) + "%\n")
    print("summary Statistics")
    print("Data Set : " + data_path)
    print()
    print()
    print("accuracy statistics for All five Experiments:")
    print(accuracy_statistics)
    print("\n")
    print("Classification Accuracy : " + str(accuracyy) + "%")
    print("\n")
    outfile_tr.close()
    outfile_ts.close()
 
main()
```
